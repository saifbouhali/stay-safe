**Introduction:**<br>
The car crash dataset under consideration is an extensive compilation of real-world incidents involving vehicular collisions. **Key Features:**<br>
1. **Incident ID:**<br>
 A unique identifier for each recorded accident, facilitating individual incident tracking and reference.

2. **DateTime:**<br>
 Timestamps for each accident, offering insights into the temporal distribution of crashes, including peak hours and days with higher incident rates.

3. **Year:**<br>
 The year in which the accident occurred, providing a temporal context for long-term trend analysis.

4. **Street Name and Cross Street:**<br>
 Geographical data indicating the locations of accidents, facilitating spatial analysis and identification of high-risk zones.

5. **Distance:**<br>
 The distance associated with each incident, providing insights into the proximity of collisions.

6. **Junction Relation:** <br>
Information about the relation to junctions, aiding in understanding the spatial context of accidents.

7. **Total Injuries and Total Fatalities:**<br>
 Numerical data on the total number of injuries and fatalities, crucial for assessing the overall impact of accidents.

8. **Injury Severity:**<br>
 Categorization of the severity of injuries sustained in accidents, ranging from minor to fatal.

9. **Collision Manner:**<br>
 Details on the manner of collision, providing insights into specific collision types (e.g., rear-end, side impact).

10. **Light Condition:**<br>
 Information on lighting conditions during accidents, contributing to understanding visibility factors.

11. **Weather and Surface Condition:**<br>
 Descriptions of weather conditions and road surface conditions at the time of each crash.

12. **Unit Type, Age, Gender, Travel Direction, Unit Action, Violation, Alcohol Use, and Drug Use for Drivers of Units One and Two:** <br>
Specific details about the vehicles involved, including the characteristics of drivers and their actions.

13. **Latitude and Longitude:**<br>
 Geographical coordinates for each incident, allowing for precise mapping and spatial analysis.
14. **Behavior:**<br>
An aggregate indicator of driving behavior, potentially derived from various behavioral variables.

<h1><font color='red'>what is gyroscope sensor:</font></h1>
A smartphone gyroscope sensor is a device integrated into modern smartphones that measures the rate of rotation or angular velocity of the device around its three axes (X, Y, and Z). It is a part of the inertial measurement unit (IMU) along with the accelerometer.

In the context of driving behavior analysis, the gyroscope can detect sudden changes in orientation, abrupt turns, and other motions that might indicate aggressive driving behavior. It's a crucial component in creating a more accurate picture of how a device (and, by extension, a vehicle) is moving in space.



<h1><font color='red'>Dataset Description:</font></h1>
Our dataset represents a comprehensive exploration of driving behavior, captured through the integration of smartphone gyroscope sensors. It includes three primary components:

***1.Acceleration (X, Y, Z axis in meters per second squared (m/s)):***
<br> This facet measures the rate of change of velocity along three spatial axes. It gauges the intensity of acceleration, crucial for understanding the dynamics of motion during driving.

***2.Rotation (X, Y, Z axis in degrees per second (/s)):***
<br> The gyroscope captures the angular velocity of the vehicle around its X, Y, and Z axes. This information provides insights into the rotational aspects of driving, helping discern patterns related to turns, twists, and overall vehicle orientation.

***3.Classification Label (SLOW, NORMAL, AGGRESSIVE):***
<br> Each data point is tagged with a classification label, indicating the driving behavior associated with that specific instance. The labels categorize driving into three classes: SLOW, NORMAL, or AGGRESSIVE. This classification serves as a key element for training predictive models aimed at assessing the safety or aggressiveness of driving behavior.

***4.Timestamp (time in seconds):***
<br> Time is a critical dimension, allowing for the temporal analysis of driving behavior. Each data entry is timestamped in seconds, facilitating the chronological understanding of events and patterns.

The integration of these parameters creates a robust dataset capable of informing predictive models.
```python
import numpy as np
from tensorflow.keras.models import load_model

# Load the saved model
loaded_model = load_model("driving_behavior_cnn.h5")

```

```python
# Load static input data from CSV file
static_data = pd.read_csv("dataset/test_motion_data.csv")  # Adjust the file path accordingly

# Assuming the static data columns are the same as your training data
# You may need to adjust this depending on the structure of your CSV file
static_inputs = static_data[['AccX', 'AccY', 'AccZ', 'GyroX', 'GyroY', 'GyroZ']].values

# Reshape the static input for prediction
static_inputs = static_inputs.reshape((static_inputs.shape[0], static_inputs.shape[1], 1))

# Make predictions for each row
pridections_array = []
for i in range(static_inputs.shape[0]):
    current_input = static_inputs[i:i+1, :, :]  # Select one row at a time
    predictions = loaded_model.predict(current_input)
    predicted_class = np.argmax(predictions)
    pridections_array.append(predicted_class)
    print(f"Prediction for Row {i + 1}: Predicted Class - {predicted_class}")
```

```python
# Define the group size
group_size = 10

behaviour_string = ['AGGRESSIVE 2 ','NORMAL 1 ','SLOW 0 ']

# Loop to print each group of 9 items
for i in range(0, len(pridections_array), group_size):
    current_group = pridections_array[i:i + group_size]
    print("Group {}: {}".format(behaviour_string[i // group_size] , current_group))
```

```python
from tensorflow.keras.models import load_model
import tensorflow as tf



# Convert the model to TensorFlow Lite format
converter = tf.lite.TFLiteConverter.from_keras_model(loaded_model)
converter.target_spec.supported_ops = [
  tf.lite.OpsSet.TFLITE_BUILTINS, # enable TensorFlow Lite ops.
  tf.lite.OpsSet.SELECT_TF_OPS # enable TensorFlow ops.
]
tflite_model = converter.convert()

# Save the TFLite model to a file
with open("driving_behavior_cnn.tflite", "wb") as f:
    f.write(tflite_model)
```
